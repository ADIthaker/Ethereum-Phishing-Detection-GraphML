## Monthly Split of Data

```python
proc_df = pd.read_csv("/content/drive/MyDrive/data/processed_graphs.csv")
```

```python
proc_df
```

```python
DATE_RANGE = (min(proc_df["date"]),max(proc_df["date"]))
DATE_RANGE
```

```python
no_of_months = 6
date_inc_range = no_of_months * 30
date_format = "%Y-%m-%d"
date_diff = datetime.datetime.strptime(DATE_RANGE[1], date_format) - datetime.datetime.strptime(DATE_RANGE[0], date_format)
date_diff.days
```

```python
curr = datetime.datetime.strptime(DATE_RANGE[0], date_format)
last_date = datetime.datetime.strptime(DATE_RANGE[1], date_format)
dates = []
get_id = lambda x: proc_df.loc[proc_df["date"] == x.strftime(date_format), "idx"].values[0]
for i in range(math.ceil(date_diff.days/date_inc_range)):
  next = curr + datetime.timedelta(days=date_inc_range)
  if next > last_date:
    next = last_date
  print(curr, next)
  dates.append((curr, next))
  if next < last_date:
    id_next = get_id(next)
    curr = next + datetime.timedelta(days=1)
    id_new_curr = get_id(curr)
    print(id_next, id_new_curr)
    if id_next == id_new_curr:
      # if the end of this interval and the start of next interval belong to same graph, include it
      # and move on another graph for next interval
      new_id = id_next+1
      # get next date value 
      curr = proc_df.loc[(proc_df["idx"] == new_id), "date"].values[0]
      # convert to datetime.datetime
      curr = datetime.datetime.strptime(curr, date_format)

  
  
```

```python
dates
```

```python
pkl_file = "/content/drive/MyDrive/data/series.pkl"
with open(pkl_file, 'wb') as handle:
  pickle.dump(len(dates), handle)
  for start, end in dates:
    start = start.strftime(date_format)
    end = end.strftime(date_format)
    idx_arr = proc_df.loc[(proc_df["date"] >= start) & (proc_df["date"] <= end), "idx"]
    pickle.dump(idx_arr, handle)
    print(idx_arr)
```
