```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'32':0.91, '64':0.98, '128':0.92}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize =(15,10))

plt.bar(courses, values, color ='maroon',
        width = 0.2)
plt.ylim([0.9, 1.0])
plt.xlabel("Embedding Size")
plt.ylabel("Accuracy")
#plt.title("Embedding sizes and classification accuracy obtained")
plt.show()
```

```python
barWidth = 0.1
fig = plt.subplots(figsize =(15,10))
 
# set height of bar
acc = [0.86,0.91, 0.969, 0.92]
prec = [0.81,0.92, 0.98, 0.93]
rec = [0.85,0.91, 0.973, 0.92]
fs = [0.83,0.91, 0.978, 0.91]

 
# Set position of bar on X axis
br1 = np.arange(len(acc))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]

 
# Make the plot
plt.bar(br1, acc, color ='r', width = barWidth,
        edgecolor ='grey', label ='Accuracy')
plt.bar(br2, prec, color ='g', width = barWidth,
        edgecolor ='grey', label ='Precision')
plt.bar(br3, rec, color ='b', width = barWidth,
        edgecolor ='grey', label ='Recall')
plt.bar(br4, fs, color ='y', width = barWidth,
        edgecolor ='grey', label ='F-Score')
 
# Adding Xticks
plt.xlabel("Embedding Size", fontweight ='bold', fontsize = 20)
plt.ylabel('Performance Values (Rounded to 2 decimal places)', fontweight ='bold', fontsize = 20)
plt.ylim([0.7, 1.0])
plt.xticks([r + 1.5*barWidth for r in range(len(acc))],
        ['16','32', '64', '128'])
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)
 
plt.legend(loc="upper left",fontsize = 20)
plt.show()
```
